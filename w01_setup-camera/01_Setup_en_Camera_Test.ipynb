## Stap 4: Detecteer Beschikbare Camera's

Voordat we beginnen, willen we weten welke camera's beschikbaar zijn op je systeem.

```python
# Functie om camera's te detecteren
def detect_cameras(max_cameras=6):
    """
    Scant alle camera indices en vindt werkende camera's
    
    Args:
        max_cameras: Maximaal aantal camera's om te testen (standaard 6)
    
    Returns:
        list: Lijst met werkende camera indices
    """
    print("Bezig met scannen van camera's...\n")
    found_cameras = []
    
    for camera_index in range(max_cameras):
        try:
            # Open camera met juiste backend voor Windows
            if sys.platform == 'win32':
                cap = cv2.VideoCapture(camera_index, cv2.CAP_DSHOW)
            else:
                cap = cv2.VideoCapture(camera_index)
            
            if cap.isOpened():
                # Probeer een frame te lezen
                ret, frame = cap.read()
                
                if ret:
                    # Haal camera eigenschappen op
                    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
                    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
                    fps = int(cap.get(cv2.CAP_PROP_FPS))
                    
                    found_cameras.append({
                        'index': camera_index,
                        'width': width,
                        'height': height,
                        'fps': fps
                    })
                    
                    print(f"✓ Camera {camera_index} gevonden!")
                    print(f"   └─ Resolutie: {width}x{height}")
                    print(f"   └─ FPS: {fps}\n")
                
                cap.release()
        
        except Exception as e:
            continue
    
    return found_cameras

# Detecteer camera's
cameras = detect_cameras()

# Toon resultaat
print("=" * 50)
if cameras:
    print(f"✓ {len(cameras)} werkende camera(s) gevonden!\n")
    
    for cam in cameras:
        print(f"Camera Index: {cam['index']}")
        print(f"   └─ Gebruik: cv2.VideoCapture({cam['index']})")
        print(f"   └─ Resolutie: {cam['width']}x{cam['height']}")
        print()
    
    # Sla de eerste camera index op voor later gebruik
    CAMERA_INDEX = cameras[0]['index']
    print(f"We gebruiken Camera {CAMERA_INDEX} voor de rest van deze notebook")
else:
    print("✗ Geen werkende camera's gevonden!")
    print("   Check TROUBLESHOOTING.md voor hulp.")
```

### Meerdere camera's?

## Stap 5: Test Je Camera

Nu gaan we een foto maken met je webcam!

```python
# Open de camera
if sys.platform == 'win32':
    cap = cv2.VideoCapture(CAMERA_INDEX, cv2.CAP_DSHOW)
else:
    cap = cv2.VideoCapture(CAMERA_INDEX)

# Check of het gelukt is
if not cap.isOpened():
    print("✗ Kan camera niet openen!")
    print("   Mogelijke oorzaken:")
    print("   - Camera wordt gebruikt door andere app (Teams, Zoom, etc.)")
    print("   - Privacy instellingen blokkeren toegang")
    print("   - Verkeerde CAMERA_INDEX")
    frame = None
else:
    print("Camera geopend! Bezig met foto maken...")
    
    # Lees een frame
    ret, frame = cap.read()
    
    if ret:
        print("✓ Foto gemaakt!")
        print(f"   └─ Frame grootte: {frame.shape}")
        print(f"   └─ {frame.shape[0]} pixels hoog")
        print(f"   └─ {frame.shape[1]} pixels breed")
        print(f"   └─ {frame.shape[2]} kleur kanalen (BGR)")
        
        # Converteer BGR naar RGB voor correcte kleuren in matplotlib
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Maak ook een grijswaarden versie voor later gebruik
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    else:
        print("✗ Kan geen frame lezen!")
        frame = None
        frame_rgb = None
        gray = None
    
    # Sluit de camera
    cap.release()
    print("\nCamera vrijgegeven")
```

**Met de code hieronder kun je de gemaakte afbeelding laten zien in het notebook**

```python
if frame_rgb is not None:
    # Toon de foto
    plt.figure(figsize=(10, 8))
    plt.imshow(frame_rgb)
    plt.title('Jouw Webcam Foto', fontsize=16)
    plt.axis('off')  # Verberg de assen
    plt.tight_layout()
    plt.show()
    
    print("Mooi! Je camera werkt perfect!")
else:
    print("✗ Geen foto beschikbaar om te tonen")
```

**Basis transformaties:**
